```python
import pandas as pd
import numpy as np
import re

# 0) CSV 읽기 (경고 억제 & 안정)
combined_df = pd.read_csv("leisure_survey_combined.csv", low_memory=False)

# 0-1) table_index 존재/형 변환
if "table_index" not in combined_df.columns:
    raise ValueError("⚠️ 'table_index' 컬럼이 CSV에 없습니다. 저장 시 df['table_index']=idx 가 포함되어야 합니다.")

# float로 저장되었을 가능성 대비
combined_df["table_index"] = pd.to_numeric(combined_df["table_index"], errors="coerce").astype("Int64")

# 사용할 표 번호
target_ids = [23, 24, 25, 28, 30, 31, 35]

# ---------- 유틸 ----------
def is_default_header(columns: pd.Index) -> bool:
    """
    컬럼이 0,1,2... 같은 RangeIndex 이거나 전부 숫자형 문자열/None에 가까우면 True.
    이미 텍스트 헤더가 들어있으면 False.
    """
    if isinstance(columns, pd.RangeIndex):
        return True
    # 예: '0','1','2' 등 숫자형 문자열이 대부분이면 기본 헤더로 판단
    num_like = 0
    for c in columns:
        if c is None:
            num_like += 1
        else:
            s = str(c).strip()
            if s.isdigit():
                num_like += 1
    return num_like >= max(2, len(columns) * 0.6)

YEAR_PATTERNS = [
    re.compile(r"^\s*(20)?(20|21|22)\s*년?\s*$"),    # 20년/21년/22년/2020/2021/2022 변형
    re.compile(r"^\s*(2020|2021|2022)\s*$"),
]

def extract_year_from_row(row_vals) -> int | float:
    for token in map(lambda x: "" if pd.isna(x) else str(x), row_vals):
        tok = token.strip()
        for pat in YEAR_PATTERNS:
            if pat.search(tok):
                # '20년' -> '20', '2022' -> '2022'
                tok_clean = tok.replace("년", "").strip()
                # '20', '21', '22'를 2020~2022로 보정
                if tok_clean in {"20","21","22"}:
                    return int("20" + tok_clean)  # 20 -> 2020, 22 -> 2022
                try:
                    return int(tok_clean)
                except:
                    pass
    return np.nan

def decide_category(row_vals) -> str:
    vals_str = " ".join([str(v) for v in row_vals if pd.notna(v)])
    if "전체" in vals_str:
        return "전체"
    if "남성" in vals_str:
        return "남성"
    if "여성" in vals_str:
        return "여성"
    # '15-19세', '20대' 등
    if re.search(r"(\d{2}\s*-\s*\d{2}\s*세)|(\d{2}\s*대)|(\d{1,2}\s*세)", vals_str):
        return "연령대"
    return "기타"

def coerce_float(x):
    """ '15.4', '15', '15,4' 등 문자열을 float로 변환, 실패 시 np.nan """
    if pd.isna(x):
        return np.nan
    s = str(x).strip().replace(",", "")
    try:
        return float(s)
    except:
        return np.nan

# ---------- 핵심: 표 정제 함수 ----------
def clean_table(df_in: pd.DataFrame, table_id: int) -> pd.DataFrame:
    # 원본 복사 & 완전 공백 행/열 제거
    df = df_in.copy()
    if "table_index" in df.columns:
        df = df.drop(columns=["table_index"])

    df = df.dropna(how="all").dropna(how="all", axis=1)
    df = df.reset_index(drop=True)

    if df.empty:
        raise ValueError("빈 데이터프레임입니다.")

    # 헤더가 기본(숫자/RangeIndex)이면 첫 행을 헤더로 사용
    if is_default_header(df.columns):
        # 첫 행이 존재할 때만
        if 0 in df.index or len(df) >= 1:
            header_row = df.iloc[0].astype(str)
            df.columns = header_row
            df = df.iloc[1:].reset_index(drop=True)

    # 이후 결측치/공백 정리
    df = df.fillna("")
    # 모든 값 문자열 정리
    df = df.applymap(lambda x: str(x).strip() if x != "" else "")

    # 메트릭 후보 컬럼: 컬럼명 자체가 공란/None이 아닌 것들
    # (필요시 더 엄격한 필터를 적용해도 됨)
    candidate_cols = [c for c in df.columns if str(c).strip() not in {"", "None"}]

    rows_out = []
    for _, row in df.iterrows():
        row_vals = [row[c] for c in candidate_cols]
        year = extract_year_from_row(row_vals)
        cat = decide_category(row_vals)

        # 각 컬럼을 돌며 숫자값만 수집 (첫번째 '인덱스성' 컬럼 제외할 필요가 있으면 여기서 제외)
        for col in candidate_cols:
            val = coerce_float(row[col])
            if not np.isnan(val):
                rows_out.append({
                    "연도": year if pd.notna(year) else np.nan,
                    "구분": cat,
                    "활동유형": str(col).replace("\n", " ").strip(),
                    "참여율(%)": val,
                    "출처표": f"표{table_id}"
                })

    return pd.DataFrame(rows_out)

# ---------- 실행 ----------
cleaned = []
for tid in target_ids:
    df_sub = combined_df.loc[combined_df["table_index"] == tid]
    try:
        df_clean = clean_table(df_sub, tid)
        if not df_clean.empty:
            cleaned.append(df_clean)
            print(f"✅ 표 {tid} 처리 완료 ({len(df_clean)} 행)")
        else:
            print(f"⚠️ 표 {tid}에서 추출된 데이터가 없습니다.")
    except Exception as e:
        print(f"⚠️ 표 {tid} 처리 실패:", e)

if cleaned:
    merged_df = pd.concat(cleaned, ignore_index=True)
    merged_df = merged_df.dropna(subset=["연도", "참여율(%)"])
    merged_df = merged_df[merged_df["참여율(%)"] > 0]
    # 저장
    out_path = "data/leisure_activity_2020_2022_clean.csv"
    merged_df.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"🎯 CSV 저장 완료 → {out_path}")
    print("총 데이터 수:", len(merged_df))
else:
    raise ValueError("❌ 병합할 데이터가 없습니다. 'table_index' 값 또는 CSV의 원본 구조를 확인하세요.")
```

# 활동유형별 변화율 계산 (2020→2022 증감률)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) 데이터 로드
df = pd.read_csv("data/leisure_activity_2020_2022_clean.csv")

# 2) 연도 정규화: (20.0/20/20년 -> 2020)
def normalize_year(x):
    s = str(x).strip()
    s = s.replace("년", "")          # "20년" -> "20"
    s = s.replace(".0", "")          # "20.0" -> "20"
    # 두 자리면 20xx로 맵
    if s in {"20","21","22"}:
        return {"20":"2020","21":"2021","22":"2022"}[s]
    # 네 자리 숫자면 그대로 사용
    if s.isdigit() and len(s) == 4:
        return s
    # 혹시 모를 값은 결측 처리
    return None

df["연도_norm"] = df["연도"].apply(normalize_year)

# 3) 필요한 연도만 필터 (2020, 2022)
df_pivot = df[df["연도_norm"].isin(["2020","2022"])].copy()

# 4) 피벗 (index=활동유형, columns=연도_norm)
pivot = df_pivot.pivot(index="활동유형", columns="연도_norm", values="참여율(%)")

# 디버깅: 현재 컬럼 확인
print("현재 피벗 컬럼:", pivot.columns.tolist())  # ['2020', '2022']가 떠야 정상

# 5) 변화율 계산 (2020→2022)
#   - 2020이 없거나 0인 경우 분모 0 방지
for col in ["2020","2022"]:
    if col not in pivot.columns:
        pivot[col] = pd.NA

pivot["변화율(%)"] = ((pivot["2022"] - pivot["2020"]) / pivot["2020"]) * 100
pivot = pivot.dropna(subset=["변화율(%)"]).sort_values("변화율(%)", ascending=False).reset_index()

# 6) 시각화 (막대그래프)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=pivot,
    y="활동유형",
    x="변화율(%)"
)
plt.title("활동유형별 참여율 변화율 (2020→2022)")
plt.xlabel("변화율(%)")
plt.ylabel("활동유형")

# 바 끝에 라벨 붙이기 (소수점 1자리)
for i, v in enumerate(pivot["변화율(%)"]):
    plt.text(v, i, f"{v:.1f}%", va="center", ha="left")

plt.tight_layout()
plt.show()
```

## 연도별 변화 라벨링

```python
# =========================================================
# 2020→2022 활동유형별 참여율 및 변화율 시각화
# =========================================================

df = pd.read_csv("data/leisure_activity_2020_2022_clean.csv")
df["연도"] = df["연도"].astype(str)
df["참여율(%)"] = pd.to_numeric(df["참여율(%)"], errors="coerce")

pivot = df.pivot_table(index="활동유형", columns="연도", values="참여율(%)")
pivot.columns = pivot.columns.astype(str)

# 🔹 연도 자동 감지
col_2020 = next((c for c in pivot.columns if "20" in c and "21" not in c and "22" not in c), None)
col_2022 = next((c for c in pivot.columns if "22" in c), None)

pivot["변화율(%)"] = ((pivot[col_2022] - pivot[col_2020]) / pivot[col_2020]) * 100
pivot = pivot.reset_index()

df_2022 = pivot[["활동유형", col_2022, "변화율(%)"]].copy()
df_2022.rename(columns={col_2022: "참여율(%)"}, inplace=True)

# =========================================================
# 시각화
# =========================================================
plt.figure(figsize=(10, 6))

colors = ["skyblue" if x >= 0 else "salmon" for x in df_2022["변화율(%)"]]
barplot = sns.barplot(
    data=df_2022,
    x="참여율(%)",
    y="활동유형",
    hue="활동유형",
    dodge=False,
    palette=dict(zip(df_2022["활동유형"], colors)),
    legend=False
)

for index, row in df_2022.iterrows():
    rate = row["참여율(%)"]
    change = row["변화율(%)"]
    sign = "+" if change >= 0 else "-"
    plt.text(rate + 0.5, index, f"{rate:.1f}% ({sign}{abs(change):.1f}%)", va="center", fontsize=10)

plt.title("2020→2022 활동유형별 참여율 및 변화율", fontsize=15, weight="bold")
plt.xlabel("2022년 참여율(%)")
plt.ylabel("활동유형")
plt.tight_layout()
plt.show()
```

```python
# =========================================================
# Leisure Activity Analysis 2020–2022 (Full Visualization)
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 0. Global Settings
# -------------------------
plt.rcParams['font.family'] = ['Arial']  # ✅ English only, no Hangul issues
plt.rcParams['axes.unicode_minus'] = False
sns.set(style="whitegrid")

# -------------------------
# 1. Load Dataset
# -------------------------
df = pd.read_csv("data/leisure_activity_2020_2022_clean.csv")

# Rename to English (avoid Korean text in any column)
df = df.rename(columns={
    "연도": "Year",
    "구분": "Category",
    "활동유형": "ActivityType",
    "참여율(%)": "ParticipationRate",
    "출처표": "SourceTable"
})

df["Year"] = df["Year"].astype(str)
df["ParticipationRate"] = pd.to_numeric(df["ParticipationRate"], errors="coerce")

# -------------------------
# 2. Pivot for Change Rate
# -------------------------
pivot = df.pivot_table(index=["ActivityType"], columns="Year", values="ParticipationRate")
pivot.columns = pivot.columns.astype(str)

col_2020 = next((c for c in pivot.columns if "20" in c and "21" not in c and "22" not in c), "2020")
col_2022 = next((c for c in pivot.columns if "22" in c), "2022")

pivot["ChangeRate(%)"] = ((pivot[col_2022] - pivot[col_2020]) / pivot[col_2020]) * 100
pivot = pivot.reset_index()

df_2022 = pivot[["ActivityType", col_2022, "ChangeRate(%)"]].rename(columns={col_2022: "ParticipationRate"})

# -------------------------
# 3. Visualization Section
# -------------------------
plt.figure(figsize=(10, 6))
print("✅ Visualization Start")

# 1️⃣ Yearly Activity Trends
plt.subplot(3, 2, 1)
sns.lineplot(data=df, x="Year", y="ParticipationRate", hue="ActivityType", marker="o")
plt.title("1) Yearly Activity Trend", fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=7)

# 2️⃣ By Gender/Age
plt.subplot(3, 2, 2)
subset = df[df["Category"].str.contains("male|female|age", case=False, na=False)]
sns.lineplot(data=subset, x="Year", y="ParticipationRate", hue="Category", style="ActivityType", marker="o")
plt.title("2) Gender/Age Trend", fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=7)

# 3️⃣ Top Activities
plt.subplot(3, 2, 3)
top_acts = df.groupby("ActivityType")["ParticipationRate"].mean().nlargest(6).index
sns.barplot(data=df[df["ActivityType"].isin(top_acts)], x="ActivityType", y="ParticipationRate", hue="Year")
plt.title("3) Top Leisure Activities", fontsize=12)

# 4️⃣ Top Activities by Gender
plt.subplot(3, 2, 4)
sub = df[df["ActivityType"].isin(top_acts) & df["Category"].str.contains("male|female", case=False, na=False)]
sns.barplot(data=sub, x="ActivityType", y="ParticipationRate", hue="Category")
plt.title("4) Top Activities by Gender", fontsize=12)

# 5️⃣ Trend Heatmap
plt.subplot(3, 2, 5)
sns.heatmap(pivot.set_index("ActivityType")[[col_2020, col_2022]], annot=True, cmap="YlGnBu")
plt.title("5) Trend Comparison (2020→2022)", fontsize=12)

plt.tight_layout()
plt.show()

# 6️⃣ Gender/Age Trend Detail
plt.figure(figsize=(10, 6))
sns.lineplot(data=subset, x="Year", y="ParticipationRate", hue="Category", marker="o")
plt.title("6) Gender and Age Trend", fontsize=12)
plt.tight_layout()
plt.show()

# 7️⃣ Sports Activities
plt.figure(figsize=(10, 6))
sports_df = df[df["ActivityType"].str.contains("sport", case=False, na=False)]
sns.barplot(data=sports_df, x="Year", y="ParticipationRate", hue="ActivityType")
plt.title("7) Sports Activity Participation", fontsize=12)
plt.tight_layout()
plt.show()

# 8️⃣ Digital & Smart Device Activities
plt.figure(figsize=(10, 6))
digital_df = df[df["ActivityType"].str.contains("mobile|internet|game|SNS|smart", case=False, na=False)]
sns.lineplot(data=digital_df, x="Year", y="ParticipationRate", hue="ActivityType", marker="o")
plt.title("8) Digital/Smart Device-based Activities", fontsize=12)
plt.tight_layout()
plt.show()

# 9️⃣ Change Rate Emphasis
plt.figure(figsize=(10, 6))
df_2022["Color"] = ["skyblue" if x >= 0 else "salmon" for x in df_2022["ChangeRate(%)"]]
sns.barplot(
    data=df_2022,
    x="ParticipationRate",
    y="ActivityType",
    hue="ActivityType",
    dodge=False,
    palette=dict(zip(df_2022["ActivityType"], df_2022["Color"])),
    legend=False
)
for index, row in df_2022.iterrows():
    plt.text(row["ParticipationRate"] + 0.3, index, f"{row['ParticipationRate']:.1f}% ({row['ChangeRate(%)']:+.1f}%)", va="center")
plt.title("9) Change Rate Emphasis (2020→2022)", fontsize=12)
plt.tight_layout()
plt.show()

# 🔟 Sorted Comparison by Gender/Age
plt.figure(figsize=(10, 6))
subset_sorted = subset[subset["Year"] == "2022"].sort_values("ParticipationRate")
sns.barplot(data=subset_sorted, x="ParticipationRate", y="Category", hue="ActivityType")
plt.title("10) Sorted Comparison by Gender/Age (2022)", fontsize=12)
plt.tight_layout()
plt.show()

print("🎯 All visualizations completed successfully!")
```
