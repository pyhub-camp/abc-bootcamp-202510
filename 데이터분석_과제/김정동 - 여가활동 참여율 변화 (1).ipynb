```python
import pandas as pd
import numpy as np
import re

# 0) CSV ì½ê¸° (ê²½ê³  ì–µì œ & ì•ˆì •)
combined_df = pd.read_csv("leisure_survey_combined.csv", low_memory=False)

# 0-1) table_index ì¡´ìž¬/í˜• ë³€í™˜
if "table_index" not in combined_df.columns:
    raise ValueError("âš ï¸ 'table_index' ì»¬ëŸ¼ì´ CSVì— ì—†ìŠµë‹ˆë‹¤. ì €ìž¥ ì‹œ df['table_index']=idx ê°€ í¬í•¨ë˜ì–´ì•¼ í•©ë‹ˆë‹¤.")

# floatë¡œ ì €ìž¥ë˜ì—ˆì„ ê°€ëŠ¥ì„± ëŒ€ë¹„
combined_df["table_index"] = pd.to_numeric(combined_df["table_index"], errors="coerce").astype("Int64")

# ì‚¬ìš©í•  í‘œ ë²ˆí˜¸
target_ids = [23, 24, 25, 28, 30, 31, 35]

# ---------- ìœ í‹¸ ----------
def is_default_header(columns: pd.Index) -> bool:
    """
    ì»¬ëŸ¼ì´ 0,1,2... ê°™ì€ RangeIndex ì´ê±°ë‚˜ ì „ë¶€ ìˆ«ìží˜• ë¬¸ìžì—´/Noneì— ê°€ê¹Œìš°ë©´ True.
    ì´ë¯¸ í…ìŠ¤íŠ¸ í—¤ë”ê°€ ë“¤ì–´ìžˆìœ¼ë©´ False.
    """
    if isinstance(columns, pd.RangeIndex):
        return True
    # ì˜ˆ: '0','1','2' ë“± ìˆ«ìží˜• ë¬¸ìžì—´ì´ ëŒ€ë¶€ë¶„ì´ë©´ ê¸°ë³¸ í—¤ë”ë¡œ íŒë‹¨
    num_like = 0
    for c in columns:
        if c is None:
            num_like += 1
        else:
            s = str(c).strip()
            if s.isdigit():
                num_like += 1
    return num_like >= max(2, len(columns) * 0.6)

YEAR_PATTERNS = [
    re.compile(r"^\s*(20)?(20|21|22)\s*ë…„?\s*$"),    # 20ë…„/21ë…„/22ë…„/2020/2021/2022 ë³€í˜•
    re.compile(r"^\s*(2020|2021|2022)\s*$"),
]

def extract_year_from_row(row_vals) -> int | float:
    for token in map(lambda x: "" if pd.isna(x) else str(x), row_vals):
        tok = token.strip()
        for pat in YEAR_PATTERNS:
            if pat.search(tok):
                # '20ë…„' -> '20', '2022' -> '2022'
                tok_clean = tok.replace("ë…„", "").strip()
                # '20', '21', '22'ë¥¼ 2020~2022ë¡œ ë³´ì •
                if tok_clean in {"20","21","22"}:
                    return int("20" + tok_clean)  # 20 -> 2020, 22 -> 2022
                try:
                    return int(tok_clean)
                except:
                    pass
    return np.nan

def decide_category(row_vals) -> str:
    vals_str = " ".join([str(v) for v in row_vals if pd.notna(v)])
    if "ì „ì²´" in vals_str:
        return "ì „ì²´"
    if "ë‚¨ì„±" in vals_str:
        return "ë‚¨ì„±"
    if "ì—¬ì„±" in vals_str:
        return "ì—¬ì„±"
    # '15-19ì„¸', '20ëŒ€' ë“±
    if re.search(r"(\d{2}\s*-\s*\d{2}\s*ì„¸)|(\d{2}\s*ëŒ€)|(\d{1,2}\s*ì„¸)", vals_str):
        return "ì—°ë ¹ëŒ€"
    return "ê¸°íƒ€"

def coerce_float(x):
    """ '15.4', '15', '15,4' ë“± ë¬¸ìžì—´ì„ floatë¡œ ë³€í™˜, ì‹¤íŒ¨ ì‹œ np.nan """
    if pd.isna(x):
        return np.nan
    s = str(x).strip().replace(",", "")
    try:
        return float(s)
    except:
        return np.nan

# ---------- í•µì‹¬: í‘œ ì •ì œ í•¨ìˆ˜ ----------
def clean_table(df_in: pd.DataFrame, table_id: int) -> pd.DataFrame:
    # ì›ë³¸ ë³µì‚¬ & ì™„ì „ ê³µë°± í–‰/ì—´ ì œê±°
    df = df_in.copy()
    if "table_index" in df.columns:
        df = df.drop(columns=["table_index"])

    df = df.dropna(how="all").dropna(how="all", axis=1)
    df = df.reset_index(drop=True)

    if df.empty:
        raise ValueError("ë¹ˆ ë°ì´í„°í”„ë ˆìž„ìž…ë‹ˆë‹¤.")

    # í—¤ë”ê°€ ê¸°ë³¸(ìˆ«ìž/RangeIndex)ì´ë©´ ì²« í–‰ì„ í—¤ë”ë¡œ ì‚¬ìš©
    if is_default_header(df.columns):
        # ì²« í–‰ì´ ì¡´ìž¬í•  ë•Œë§Œ
        if 0 in df.index or len(df) >= 1:
            header_row = df.iloc[0].astype(str)
            df.columns = header_row
            df = df.iloc[1:].reset_index(drop=True)

    # ì´í›„ ê²°ì¸¡ì¹˜/ê³µë°± ì •ë¦¬
    df = df.fillna("")
    # ëª¨ë“  ê°’ ë¬¸ìžì—´ ì •ë¦¬
    df = df.applymap(lambda x: str(x).strip() if x != "" else "")

    # ë©”íŠ¸ë¦­ í›„ë³´ ì»¬ëŸ¼: ì»¬ëŸ¼ëª… ìžì²´ê°€ ê³µëž€/Noneì´ ì•„ë‹Œ ê²ƒë“¤
    # (í•„ìš”ì‹œ ë” ì—„ê²©í•œ í•„í„°ë¥¼ ì ìš©í•´ë„ ë¨)
    candidate_cols = [c for c in df.columns if str(c).strip() not in {"", "None"}]

    rows_out = []
    for _, row in df.iterrows():
        row_vals = [row[c] for c in candidate_cols]
        year = extract_year_from_row(row_vals)
        cat = decide_category(row_vals)

        # ê° ì»¬ëŸ¼ì„ ëŒë©° ìˆ«ìžê°’ë§Œ ìˆ˜ì§‘ (ì²«ë²ˆì§¸ 'ì¸ë±ìŠ¤ì„±' ì»¬ëŸ¼ ì œì™¸í•  í•„ìš”ê°€ ìžˆìœ¼ë©´ ì—¬ê¸°ì„œ ì œì™¸)
        for col in candidate_cols:
            val = coerce_float(row[col])
            if not np.isnan(val):
                rows_out.append({
                    "ì—°ë„": year if pd.notna(year) else np.nan,
                    "êµ¬ë¶„": cat,
                    "í™œë™ìœ í˜•": str(col).replace("\n", " ").strip(),
                    "ì°¸ì—¬ìœ¨(%)": val,
                    "ì¶œì²˜í‘œ": f"í‘œ{table_id}"
                })

    return pd.DataFrame(rows_out)

# ---------- ì‹¤í–‰ ----------
cleaned = []
for tid in target_ids:
    df_sub = combined_df.loc[combined_df["table_index"] == tid]
    try:
        df_clean = clean_table(df_sub, tid)
        if not df_clean.empty:
            cleaned.append(df_clean)
            print(f"âœ… í‘œ {tid} ì²˜ë¦¬ ì™„ë£Œ ({len(df_clean)} í–‰)")
        else:
            print(f"âš ï¸ í‘œ {tid}ì—ì„œ ì¶”ì¶œëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
    except Exception as e:
        print(f"âš ï¸ í‘œ {tid} ì²˜ë¦¬ ì‹¤íŒ¨:", e)

if cleaned:
    merged_df = pd.concat(cleaned, ignore_index=True)
    merged_df = merged_df.dropna(subset=["ì—°ë„", "ì°¸ì—¬ìœ¨(%)"])
    merged_df = merged_df[merged_df["ì°¸ì—¬ìœ¨(%)"] > 0]
    # ì €ìž¥
    out_path = "data/leisure_activity_2020_2022_clean.csv"
    merged_df.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"ðŸŽ¯ CSV ì €ìž¥ ì™„ë£Œ â†’ {out_path}")
    print("ì´ ë°ì´í„° ìˆ˜:", len(merged_df))
else:
    raise ValueError("âŒ ë³‘í•©í•  ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤. 'table_index' ê°’ ë˜ëŠ” CSVì˜ ì›ë³¸ êµ¬ì¡°ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
```

```python
import pandas as pd
import numpy as np

# ëŒ€ìƒ í‘œ ì¸ë±ìŠ¤
target_tables = [tables[23], tables[24], tables[25], tables[28], tables[30], tables[31], tables[35]]
cleaned = []

def clean_table(df, table_id):
    # í—¤ë” í–‰ ê°€ì ¸ì˜¤ê¸°
    df.columns = df.iloc[0].astype(str)
    df = df.drop(0).reset_index(drop=True)
    df = df.fillna("")  # ê²°ì¸¡ì¹˜ ì²˜ë¦¬

    rows = []
    for _, row in df.iterrows():
        year = None
        # ì—°ë„ ì¶”ì¶œ
        for token in row.astype(str).values:
            if any(x in token for x in ["20ë…„", "21ë…„", "22ë…„", "2020", "2021", "2022"]):
                year = token.replace("ë…„", "").strip()
        # ì „ì²´/ì„±ë³„/ì—°ë ¹ëŒ€ ë“± êµ¬ë¶„
        if "ì „ì²´" in row.values: cat = "ì „ì²´"
        elif "ë‚¨ì„±" in row.values: cat = "ë‚¨ì„±"
        elif "ì—¬ì„±" in row.values: cat = "ì—¬ì„±"
        elif any(x in str(row.values) for x in ["ëŒ€", "ì„¸"]): cat = "ì—°ë ¹ëŒ€"
        else: cat = "ê¸°íƒ€"
        
        # ë°ì´í„° íŽ¼ì¹˜ê¸°
        for col in df.columns[1:]:
            val = row[col]
            if isinstance(val, (int, float)) or str(val).replace(".", "", 1).isdigit():
                rows.append({
                    "ì—°ë„": int(year) if year else np.nan,
                    "êµ¬ë¶„": cat,
                    "í™œë™ìœ í˜•": col.replace("\n", " ").strip(),
                    "ì°¸ì—¬ìœ¨(%)": float(val),
                    "ì¶œì²˜í‘œ": f"í‘œ{table_id}"
                })
    return pd.DataFrame(rows)

# í‘œë³„ ì²˜ë¦¬
for i, t in zip([23,24,25,28,30,31,35], target_tables):
    try:
        df_clean = clean_table(t, i)
        cleaned.append(df_clean)
        print(f"âœ… í‘œ {i} ì²˜ë¦¬ ì™„ë£Œ ({len(df_clean)} í–‰)")
    except Exception as e:
        print(f"âš ï¸ í‘œ {i} ì²˜ë¦¬ ì‹¤íŒ¨:", e)

# ë³‘í•©
merged_df = pd.concat(cleaned, ignore_index=True)
merged_df = merged_df.dropna(subset=["ì—°ë„", "ì°¸ì—¬ìœ¨(%)"])
merged_df = merged_df[merged_df["ì°¸ì—¬ìœ¨(%)"] > 0]

merged_df.to_csv("data/leisure_activity_2020_2022_clean.csv", index=False, encoding="utf-8-sig")
print("ðŸŽ¯ CSV ì €ìž¥ ì™„ë£Œ â†’ data/leisure_activity_2020_2022_clean.csv")
print("ì´ ë°ì´í„° ìˆ˜:", len(merged_df))
merged_df
```

# í™œë™ìœ í˜•ë³„ ë³€í™”ìœ¨ ê³„ì‚° (2020â†’2022 ì¦ê°ë¥ )

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) ë°ì´í„° ë¡œë“œ
df = pd.read_csv("data/leisure_activity_2020_2022_clean.csv")

# 2) ì—°ë„ ì •ê·œí™”: (20.0/20/20ë…„ -> 2020)
def normalize_year(x):
    s = str(x).strip()
    s = s.replace("ë…„", "")          # "20ë…„" -> "20"
    s = s.replace(".0", "")          # "20.0" -> "20"
    # ë‘ ìžë¦¬ë©´ 20xxë¡œ ë§µ
    if s in {"20","21","22"}:
        return {"20":"2020","21":"2021","22":"2022"}[s]
    # ë„¤ ìžë¦¬ ìˆ«ìžë©´ ê·¸ëŒ€ë¡œ ì‚¬ìš©
    if s.isdigit() and len(s) == 4:
        return s
    # í˜¹ì‹œ ëª¨ë¥¼ ê°’ì€ ê²°ì¸¡ ì²˜ë¦¬
    return None

df["ì—°ë„_norm"] = df["ì—°ë„"].apply(normalize_year)

# 3) í•„ìš”í•œ ì—°ë„ë§Œ í•„í„° (2020, 2022)
df_pivot = df[df["ì—°ë„_norm"].isin(["2020","2022"])].copy()

# 4) í”¼ë²— (index=í™œë™ìœ í˜•, columns=ì—°ë„_norm)
pivot = df_pivot.pivot(index="í™œë™ìœ í˜•", columns="ì—°ë„_norm", values="ì°¸ì—¬ìœ¨(%)")

# ë””ë²„ê¹…: í˜„ìž¬ ì»¬ëŸ¼ í™•ì¸
print("í˜„ìž¬ í”¼ë²— ì»¬ëŸ¼:", pivot.columns.tolist())  # ['2020', '2022']ê°€ ë– ì•¼ ì •ìƒ

# 5) ë³€í™”ìœ¨ ê³„ì‚° (2020â†’2022)
#   - 2020ì´ ì—†ê±°ë‚˜ 0ì¸ ê²½ìš° ë¶„ëª¨ 0 ë°©ì§€
for col in ["2020","2022"]:
    if col not in pivot.columns:
        pivot[col] = pd.NA

pivot["ë³€í™”ìœ¨(%)"] = ((pivot["2022"] - pivot["2020"]) / pivot["2020"]) * 100
pivot = pivot.dropna(subset=["ë³€í™”ìœ¨(%)"]).sort_values("ë³€í™”ìœ¨(%)", ascending=False).reset_index()

# 6) ì‹œê°í™” (ë§‰ëŒ€ê·¸ëž˜í”„)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=pivot,
    y="í™œë™ìœ í˜•",
    x="ë³€í™”ìœ¨(%)"
)
plt.title("í™œë™ìœ í˜•ë³„ ì°¸ì—¬ìœ¨ ë³€í™”ìœ¨ (2020â†’2022)")
plt.xlabel("ë³€í™”ìœ¨(%)")
plt.ylabel("í™œë™ìœ í˜•")

# ë°” ëì— ë¼ë²¨ ë¶™ì´ê¸° (ì†Œìˆ˜ì  1ìžë¦¬)
for i, v in enumerate(pivot["ë³€í™”ìœ¨(%)"]):
    plt.text(v, i, f"{v:.1f}%", va="center", ha="left")

plt.tight_layout()
plt.show()
```

```python
# =========================================================
# Leisure Activity Analysis 2020â€“2022 (Full Visualization)
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 0. Global Settings
# -------------------------
plt.rcParams['font.family'] = ['Arial']  # âœ… English only, no Hangul issues
plt.rcParams['axes.unicode_minus'] = False
sns.set(style="whitegrid")

# -------------------------
# 1. Load Dataset
# -------------------------
df = pd.read_csv("data/leisure_activity_2020_2022_clean.csv")

# Rename to English (avoid Korean text in any column)
df = df.rename(columns={
    "ì—°ë„": "Year",
    "êµ¬ë¶„": "Category",
    "í™œë™ìœ í˜•": "ActivityType",
    "ì°¸ì—¬ìœ¨(%)": "ParticipationRate",
    "ì¶œì²˜í‘œ": "SourceTable"
})

df["Year"] = df["Year"].astype(str)
df["ParticipationRate"] = pd.to_numeric(df["ParticipationRate"], errors="coerce")

# -------------------------
# 2. Pivot for Change Rate
# -------------------------
pivot = df.pivot_table(index=["ActivityType"], columns="Year", values="ParticipationRate")
pivot.columns = pivot.columns.astype(str)

col_2020 = next((c for c in pivot.columns if "20" in c and "21" not in c and "22" not in c), "2020")
col_2022 = next((c for c in pivot.columns if "22" in c), "2022")

pivot["ChangeRate(%)"] = ((pivot[col_2022] - pivot[col_2020]) / pivot[col_2020]) * 100
pivot = pivot.reset_index()

df_2022 = pivot[["ActivityType", col_2022, "ChangeRate(%)"]].rename(columns={col_2022: "ParticipationRate"})

# -------------------------
# 3. Visualization Section
# -------------------------
plt.figure(figsize=(10, 6))
print("âœ… Visualization Start")

# 1ï¸âƒ£ Yearly Activity Trends
plt.subplot(3, 2, 1)
sns.lineplot(data=df, x="Year", y="ParticipationRate", hue="ActivityType", marker="o")
plt.title("1) Yearly Activity Trend", fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=7)

# 2ï¸âƒ£ By Gender/Age
plt.subplot(3, 2, 2)
subset = df[df["Category"].str.contains("male|female|age", case=False, na=False)]
sns.lineplot(data=subset, x="Year", y="ParticipationRate", hue="Category", style="ActivityType", marker="o")
plt.title("2) Gender/Age Trend", fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=7)

# 3ï¸âƒ£ Top Activities
plt.subplot(3, 2, 3)
top_acts = df.groupby("ActivityType")["ParticipationRate"].mean().nlargest(6).index
sns.barplot(data=df[df["ActivityType"].isin(top_acts)], x="ActivityType", y="ParticipationRate", hue="Year")
plt.title("3) Top Leisure Activities", fontsize=12)

# 4ï¸âƒ£ Top Activities by Gender
plt.subplot(3, 2, 4)
sub = df[df["ActivityType"].isin(top_acts) & df["Category"].str.contains("male|female", case=False, na=False)]
sns.barplot(data=sub, x="ActivityType", y="ParticipationRate", hue="Category")
plt.title("4) Top Activities by Gender", fontsize=12)

# 5ï¸âƒ£ Trend Heatmap
plt.subplot(3, 2, 5)
sns.heatmap(pivot.set_index("ActivityType")[[col_2020, col_2022]], annot=True, cmap="YlGnBu")
plt.title("5) Trend Comparison (2020â†’2022)", fontsize=12)

plt.tight_layout()
plt.show()

# 6ï¸âƒ£ Gender/Age Trend Detail
plt.figure(figsize=(10, 6))
sns.lineplot(data=subset, x="Year", y="ParticipationRate", hue="Category", marker="o")
plt.title("6) Gender and Age Trend", fontsize=12)
plt.tight_layout()
plt.show()

# 7ï¸âƒ£ Sports Activities
plt.figure(figsize=(10, 6))
sports_df = df[df["ActivityType"].str.contains("sport", case=False, na=False)]
sns.barplot(data=sports_df, x="Year", y="ParticipationRate", hue="ActivityType")
plt.title("7) Sports Activity Participation", fontsize=12)
plt.tight_layout()
plt.show()

# 8ï¸âƒ£ Digital & Smart Device Activities
plt.figure(figsize=(10, 6))
digital_df = df[df["ActivityType"].str.contains("mobile|internet|game|SNS|smart", case=False, na=False)]
sns.lineplot(data=digital_df, x="Year", y="ParticipationRate", hue="ActivityType", marker="o")
plt.title("8) Digital/Smart Device-based Activities", fontsize=12)
plt.tight_layout()
plt.show()

# 9ï¸âƒ£ Change Rate Emphasis
plt.figure(figsize=(10, 6))
df_2022["Color"] = ["skyblue" if x >= 0 else "salmon" for x in df_2022["ChangeRate(%)"]]
sns.barplot(
    data=df_2022,
    x="ParticipationRate",
    y="ActivityType",
    hue="ActivityType",
    dodge=False,
    palette=dict(zip(df_2022["ActivityType"], df_2022["Color"])),
    legend=False
)
for index, row in df_2022.iterrows():
    plt.text(row["ParticipationRate"] + 0.3, index, f"{row['ParticipationRate']:.1f}% ({row['ChangeRate(%)']:+.1f}%)", va="center")
plt.title("9) Change Rate Emphasis (2020â†’2022)", fontsize=12)
plt.tight_layout()
plt.show()

# ðŸ”Ÿ Sorted Comparison by Gender/Age
plt.figure(figsize=(10, 6))
subset_sorted = subset[subset["Year"] == "2022"].sort_values("ParticipationRate")
sns.barplot(data=subset_sorted, x="ParticipationRate", y="Category", hue="ActivityType")
plt.title("10) Sorted Comparison by Gender/Age (2022)", fontsize=12)
plt.tight_layout()
plt.show()

print("ðŸŽ¯ All visualizations completed successfully!")
```
