## 연도별 상위 5개 영화의 관객 수 평균 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# ===== 한글 폰트 설정 =====
try:
    font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows 맑은 고딕
    font_name = font_manager.FontProperties(fname=font_path).get_name()
    rc('font', family=font_name)
except Exception:
    pass
plt.rcParams['axes.unicode_minus'] = False

# ===== 데이터 불러오기 =====
df_2014 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2014.csv")
df_2015 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2015.csv")
df_2016 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2016.csv")
df_2017 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2017.csv")
df_2018 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2018.csv")
df_2019 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2019.csv")
df_2020 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2020.csv")
df_2021 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2021.csv")
df_2022 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2022.csv")
df_2023 = pd.read_csv(r"C:\work\텐트메이커1기\데이터분석\data\kobis_2023.csv")

# ===== 유틸: 관객수 컬럼 찾기 & 숫자형으로 정리 =====
def find_audience_col(df):
    if '관객수' in df.columns:
        return '관객수'
    for c in df.columns:
        if '관객' in c and ('수' in c or '합' in c or '누적' in c or '총' in c):
            return c
    return None

def to_numeric_series(s):
    s = s.astype(str).str.replace(',', '', regex=False)
    s = s.str.replace(r'[^\d\.\-]', '', regex=True)
    return pd.to_numeric(s, errors='coerce')

def top5_mean_audience(df):
    col = find_audience_col(df)
    if col is None:
        raise ValueError("관객수 컬럼을 찾지 못했습니다. CSV의 컬럼명을 확인하세요.")
    num = to_numeric_series(df[col])
    return num.dropna().nlargest(10).mean()

# ===== 각 연도별 상위 5개 평균 계산 =====
avg_top5 = {
    '2014': top5_mean_audience(df_2014),
    '2015': top5_mean_audience(df_2015),
    '2016': top5_mean_audience(df_2016),
    '2017': top5_mean_audience(df_2017),
    '2018': top5_mean_audience(df_2018),
    '2019': top5_mean_audience(df_2019),
    '2020': top5_mean_audience(df_2020),
    '2021': top5_mean_audience(df_2021),
    '2022': top5_mean_audience(df_2022),
    '2023': top5_mean_audience(df_2023),
}

# ===== 시각화 =====
years = list(avg_top5.keys())
values = list(avg_top5.values())

plt.figure(figsize=(10, 6))

# 🎨 막대그래프 색상 지정
colors = ["skyblue","skyblue","pink","pink","pink","pink", "red", "red", "pink", "pink"]
bars = plt.bar(years, values, color=colors, edgecolor='black', label='평균 관객수')

# 📈 추세선 추가
plt.plot(years, values, color='black', marker='o', linewidth=1, label='추세선')

# 제목 / 라벨 / 격자
plt.title('연도별 상위 10개 영화 관객수 평균 (2014–2023)', fontsize=15, fontweight='bold')
plt.xlabel('연도', fontsize=12)
plt.ylabel('평균 관객수', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 막대 위 값 표시 (천단위 구분)
for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{val:,.0f}',
             ha='center', va='bottom', fontsize=11)

# 범례 추가
plt.legend()

plt.tight_layout()
plt.show()
```

## 연도별 1위 영화 매출액 비교 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import textwrap

# ===== 한글 폰트 설정 =====
try:
    font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows: 맑은 고딕
    font_name = font_manager.FontProperties(fname=font_path).get_name()
    rc('font', family=font_name)
except Exception:
    pass
plt.rcParams['axes.unicode_minus'] = False

# ===== 매출액 숫자형 변환 함수 =====
def to_numeric_series(s):
    s = s.astype(str).str.replace(',', '', regex=False)
    s = s.str.replace(r'[^\d\.\-]', '', regex=True)
    return pd.to_numeric(s, errors='coerce')

# ===== 각 연도의 1위 영화 정보(제목, 매출액) 추출 함수 =====
def get_top1_info(df):
    df = df.copy()
    if '순위' in df.columns:
        df = df.sort_values(by='순위', ascending=True)
    if '매출액' not in df.columns or '영화명' not in df.columns:
        raise ValueError("필수 컬럼(매출액, 영화명)이 없습니다.")
    df['매출액(숫자)'] = to_numeric_series(df['매출액'])
    top_row = df.iloc[0]
    return top_row['영화명'], top_row['매출액(숫자)']

# ===== 연도별 1위 영화 제목과 매출액 추출 =====
top1_data = {
    '2014': get_top1_info(df_2014),
    '2015': get_top1_info(df_2015),
    '2016': get_top1_info(df_2016),
    '2017': get_top1_info(df_2017),
    '2018': get_top1_info(df_2018),
    '2019': get_top1_info(df_2019),
    '2020': get_top1_info(df_2020),
    '2021': get_top1_info(df_2021),
    '2022': get_top1_info(df_2022),
    '2023': get_top1_info(df_2023),
}

years  = list(top1_data.keys())
titles = [v[0] for v in top1_data.values()]
values = [v[1] for v in top1_data.values()]

# ===== 색상 설정 (원하는 연도 강조 가능: 2020 강조 예시) =====
base_colors = {
    '2014': 'blue',
    '2015': 'blue',
    '2016': 'blue',
    '2017': 'blue',
    '2018': 'blue',
    '2019': 'blue',
    '2020': 'red',  # 강조 색
    '2021': 'red',
    '2022': 'blue',
    '2023': 'blue'
}
colors = [base_colors.get(y, '#6fa8dc') for y in years]

# ===== x축 레이블: "연도\n(줄바꿈)감싸진 영화제목" =====
def wrap_title(t, width=12):
    return textwrap.fill(str(t), width=width)

xlabels = [f"{y}\n{wrap_title(t)}" for y, t in zip(years, titles)]

# ===== 시각화 =====
plt.figure(figsize=(12, 7))
bars = plt.bar(years, values, color=colors, edgecolor='black')

plt.title('연도별 1위 영화 매출액 비교 (2014–2023)', fontsize=16, fontweight='bold')
plt.xlabel('연도 / 1위 영화 제목', fontsize=12)
plt.ylabel('매출액 (원)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 막대 위 금액 표기
for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{val:,.0f}', ha='center', va='bottom', fontsize=11)

# === 겹침 해결 포인트 ===
# 1) x축 틱 라벨을 연도+제목(줄바꿈)으로 교체
plt.xticks(ticks=range(len(years)), labels=xlabels, rotation=0, ha='center')

# 2) 하단 여백 늘려서 레이블이 잘리거나 겹치지 않게
plt.subplots_adjust(bottom=0.28)

plt.tight_layout()
plt.show()
```
