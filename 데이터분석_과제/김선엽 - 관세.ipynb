```python
# 1️⃣ 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 2️⃣ 데이터 불러오기
df = pd.read_csv("data/관세청 데이터/세수 총괄_20251030.csv")

# 3️⃣ '총계' 행 제거 및 '기간'을 숫자형으로 변환
df = df[df["기간"] != "총계"].copy()
df["기간"] = df["기간"].astype(int)

# 4️⃣ 시각화 스타일 설정
plt.style.use('seaborn-v0_8')
plt.rcParams['font.family'] = 'Malgun Gothic'  # (Mac은 'AppleGothic' 사용)
plt.rcParams['axes.unicode_minus'] = False

# 5️⃣ 세수 수입 변화 (합계)
plt.figure(figsize=(10, 5))
plt.plot(df["기간"], df["합계"], marker='o', linewidth=2, color='royalblue')
plt.title("연도별 세수 수입 변화", fontsize=14)
plt.xlabel("연도")
plt.ylabel("세수 합계 (억 원)")
plt.grid(True)
plt.show()

# 6️⃣ 관세징수율 변화
plt.figure(figsize=(10, 5))
plt.plot(df["기간"], df["관세징수율"], marker='s', color='orange', linewidth=2)
plt.title("연도별 관세징수율 변화 (%)", fontsize=14)
plt.xlabel("연도")
plt.ylabel("관세징수율 (%)")
plt.grid(True)
plt.show()

# 7️⃣ 부가가치세 변화
plt.figure(figsize=(10, 5))
plt.plot(df["기간"], df["부가세"],  marker='s', color='seagreen')  #bar
plt.title("연도별 부가가치세 변화", fontsize=14)
plt.xlabel("연도")
plt.ylabel("부가세 (억 원)")
plt.grid(True)
plt.show()

# 8️⃣ 수입액 변화
plt.figure(figsize=(10, 5))
plt.plot(df["기간"], df["수입액"], marker='D', color='purple', linewidth=2)
plt.title("연도별 수입액 변화", fontsize=14)
plt.xlabel("연도")
plt.ylabel("수입액 (억 원)")
plt.grid(True)
plt.show()
```
